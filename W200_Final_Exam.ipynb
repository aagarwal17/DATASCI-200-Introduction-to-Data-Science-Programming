(For example 31 Jan 2018 was $1200 as shown in `data` below).

The initial investment is made on 1 Jan 2018 and was for $1000 (labelled 'Initial' in the `data` below).

Fill in the Python function (named `bitcoin_invest` below) to take such a record of any length (the below `data` is only a sample but the full data will be in the same format), and return a dataframe comprising a row for each month with columns for date, balance, and return. **Print out/display this dataframe.**

Also, visualize the record as two vertically arranged plots (there are a few ways to do this, keep it consistent over both plots). 
- The top plot should show a line plot of balance vs. month 
- The bottom plot should show a bar plot of return vs. month, with a black horizontal line at return=0, and bars color-coded such that positive returns are green and negative returns are red.  
- The two plots' horizontal axes should align. Demonstrate that your code works by applying it to `data`.

Some definitions: <br>
- The gain for each period is the end balance minus the start balance. 
- The growth factor for each period is the end balance divided by the start balance.  
- The return for each period is the growth factor minus 1.

````python
data = ({"Initial": 1000}, {"Jan 2018":1200},{"Feb 2018":1400},{"Mar 2018":700},{"Apr 2018":800},{"May 2018":500})
data
````

````python
# 3) Your code here
import pandas as pd
import matplotlib.pyplot as plt

def bitcoin_invest(monthly_data):
    theData = [list(d.items())[0] for d in monthly_data]
    dataInDf = pd.DataFrame(theData, columns=['date', 'balance'])
    dataInDf['return'] = dataInDf['balance']/dataInDf['balance'].shift(1) - 1
    print(dataInDf)
    
    plt.figure(figsize=(16, 12))

    # Line plot for balance vs. month
    plt.subplot(2, 1, 1)
    plt.plot(dataInDf['date'], dataInDf['balance'])
    plt.title('Balance vs. Month')
    plt.xlabel('Month')
    plt.ylabel('Balance')
    
    # Bar plot for return vs. month
    plt.subplot(2, 1, 2)
    plt.bar(dataInDf['date'], dataInDf['return'], color=['green' if retVal > 0 else 'red' for retVal in dataInDf['return']])
    plt.axhline(0, color='black')
    plt.title('Return vs. Month')
    plt.xlabel('Month')
    plt.ylabel('Return')
    
    plt.show()

bitcoin_invest(data)
````

### VCF file description (Summarized from version 4.1)

```
* The VCF specification:

VCF is a text file format which contains meta-information lines, a header line, and then data lines each containing information about a position in the genome. The format also can contain genotype information on samples for each position.

* Fixed fields:

There are 8 fixed fields per record. All data lines are **tab-delimited**. In all cases, missing values are specified with a dot (.). 

1. CHROM - chromosome number
2. POS - position DNA nuceleotide count (bases) along the chromosome
3. ID - The unique identifier for each mutation
4. REF - reference base(s)
5. ALT - alternate base(s)
6. FILTER - filter status
7. QUAL - quality
8. INFO - a semicolon-separated series of keys with values in the format: <key>=<data>

```
### Applicable INFO field specifications

```
GENEINFO = <Gene name>
CLNSIG =  <Clinical significance>
CLNDN = <Disease name>
```

````python
# 4) Your code here - can use as many code blocks as you would like
import pandas as pd
import numpy as np
import io

# Read in Data using code similar to provided in github link:
def read_vcf(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not l.startswith('#')]
    return pd.read_csv(
        io.StringIO(''.join(lines)),
        sep='\t'
    )

df = read_vcf('../data/clinvar_final.txt')
df['POS'] = df['POS'].str.replace('O', '0')
df = df.astype({'CHROM': int, 'POS': int,'ID': str,'REF': str,'ALT': str,'FILTER': str,'QUAL': str,'INFO': str})
df
````

````python
# Datatypes make sense and no apparant nulls
df.info()
````

````python
# Remove rows with missing values
df.dropna(how='all', inplace=True)
# Remove rows with missing values in the 'INFO' column
df = df[df['INFO'].notna()]
df
````

````python
# Create Columns for GENEINFO, CLNSIG, and CLNDN:
def extract_col(theColVal, theColumn):
    INFO_split = theColVal.split(';')
    for each_part in INFO_split:
        if each_part.startswith(theColumn):
            return each_part.split('=')[1]
    return 'Not_Given'

df['GENEINFO'] = df['INFO'].apply(lambda x: extract_col(x, 'GENEINFO'))
df['CLNSIG'] = df['INFO'].apply(lambda x: extract_col(x, 'CLNSIG'))
df['CLNDN'] = df['INFO'].apply(lambda x: extract_col(x, 'CLNDN'))
df
````

````python
# Replace nan's with 'Not_Given'
df = df.replace(np.nan, 'Not_Given')
# Replace empty strings with 'Not_Given'
df = df.replace('', 'Not_Given')
# Replace '.' with 'Not_Given'
df = df.replace('.', 'Not_Given')
df
````

````python
#If the value in CLNSIG starts with a number, set the value in new column CLNSIG_words to be everything after ':'. Otherwise, set it to the value in CLNSIG.
df['CLNSIG_words'] = df['CLNSIG'].apply(lambda x: x.split(':')[1] if x[0].isdigit() else x)
# Remove any remaining numbers:
df['CLNSIG_words'] = df['CLNSIG_words'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
# Remove any remaining whitespace:
df['CLNSIG_words'] = df['CLNSIG_words'].apply(lambda x: x.strip())
# Remove vertical bars:
df['CLNSIG_words'] = df['CLNSIG_words'].apply(lambda x: x.replace('|', ''))
# Replace underscores with spaces:
df['CLNSIG_words'] = df['CLNSIG_words'].apply(lambda x: x.replace('_', ' '))
# Sentence Case:
df['CLNSIG_words'] = df['CLNSIG_words'].apply(lambda x: x.title())
````

Notes:
- Protective:  reduces the risk of disease. Protective mutations disable a gene but create a benefit rather than a problem. We will assume this means NOT DANGEROUS. 
- Pathogenic: a genetic change that increases an individual's risk of developing a disease or disorder. We will assume this means DANGEROUS
- Benign: a variant that does not cause disease and has no impact on health. We will assume this means NOT DANGEROUS. 
- Risk Factor: variant increases the likelihood of developing a disease. We will assume this means DANGEROUS
- Affects: Not enough info, affects can be beneficial, neutral, or harmful. We will have to ignore these cases (label them as not dangerous)
- Association: Not enough info, we will have to ignore these cases (label them as not dangerous)
- Conflicting: Hard to say for certain; we will ignore these cases (label them as not dangerous)
- Other/Not Given/Uncertain Signficance: not enough info, we do not include these cases as dangerous due to lack of info

Overall, if CLNSIG_words column contains Pathogenic or Risk Factor, label it as Dangerous.
