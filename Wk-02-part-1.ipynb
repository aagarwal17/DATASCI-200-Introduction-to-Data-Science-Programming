<h2 style="color:red;">OBJECTS - Part 2</h2>
<p>An <b>object</b> in computing consists of 2 parts - the properties (or facets or attributes or state) of an object and what that object can <i>do</i>, what kind of action (function or procedure) it does, called the <b>method</b>.  Wrapping both the attributes + methods into a single code block is called <b>object</b>.
    </p>
    <ol>
    <li>the wrapping of properties + methods into a single entity is called <b>encapsulation</b></li>
    <li>Once defined, an object can be a parent to other objects, reused and new properties/methods added to it.  This is <b>inheritance</b>.</li>
    <li>We can define an object with one name, but pass a different number of arguments to it; this is <b>polymorphism</b>.</li>
    </ol>
    <img src="images/object-2.jpg" style="width:270px;">
<p>Python&rsquo;s polymorphism is different than other languages&rsquo; versions. The <code>print()</code> command is a good example: all of these will work, without causing an error: <code>print()</code>, <code>print("hello")</code>, <code>print("winner is ", 25)</code>.  Each has a different number of <b>parameters</b>.  When an object is defined by us, the data passed to it are called <b>arguments</b>; when expressed in code, our objects are called <b>classes</b> and the parameters are called <b>arguments</b>.</p>

<h2 style="color:red;">OBJECTS - Part 3</h2>
<p>The python language consists of an <b>object hierarchy</b>.  This means all of the datatypes in python and their behaviors and attributes are encapsulated in objects; and each object has parent object ... ultimately working our way to the <b>root</b> of the hierarchy - an object actually called Object.</p>
<p>By writing python code, our scripts <i>inherit</i> all of python&rsquo;s object hierarchy.  That means <i>our</i> strings inherit all the properties and methods or python&rsquo;s String object.
</p>
<h3>Constructing an object</h3>
<p>Just by declaring a variable and instantiating it, we shortcut a <b>constructor</b>.  For instance:
<pre>
my_cat = "Kitty"
</pre>
means we declared a variable of type String and assigned it a value of 'Kitty'.  This is the equivalent of Java's <code>String my_cat = new String("Kitty")</code> (a pretty verbose constructor).</p>

<h3>What our objects inherit</h3>
<p>Given the string example of "my_kitty", we inherit all the String methods.  Here are some examples (ask yourself, where they come from and how you can identify code you didn&rsquo;t write yourself):</p>
<pre>
my_cat = "Kitty"
my_cat.upper()     <i>outputs KITTY</i>
my_cat.isascii()   <i>returns True - the characters are in the ASCII encoding set</i>
type(my_cat)       <i>returns &lt;class: 'str'&gt;</i>
id(my_cat)         <i>returns a value, like 932812928, a unique identifier for that variable</i>
my_cat.__class__   <i>returns &lt;class: 'str'&gt;</i>
</pre>
<p>What is that __ and the dot?</p>

<h3>Where our objects inherit... the Python Object Hierarchy</h3>
<p>Notice that some of the methods are separated from the variable by a . (dot).  Anything to the right of the dot (with the ()) means that command is <i>built-into</i> the object (on the left).  Since "my_cat" is a String object, the method .upper() belongs to the parent in python, a String Object.</p>

<h3>Accessing attributes and methods inside an Object</h3>
<p>Terms starting with a double-underscore, or starting &amp; ending with the double underscore (called a <b>dunder</b>), come from Python's Objects.  We use these <b>magic methods</b> to access info about the object and use them to change the default behavior of objects.  [More on this later this term.]

<h3>Turning our Object Definition into a Class</h3>
<p>Here we want to make a first attempt at defining a Book class.  The convention is to use initial capitals for all classes.</p>
<pre>
class Book(no_of_pages, author, title, subject, ID_No, owner="Jane Smith")
    this.no_of_pages = no_of_pages
    this.author = author
    # and so on ... 
</pre>
<p>Having defined a Book class, let's construct one:</p>
<pre>
my_Hamlet_Book = Book(75, "Shakespeare", "Hamlet", "Drama", "PE399.2", "Bill Ney")
</pre>
<p>Notice the syntax.  There&rsquo;s a one-to-one relationship between the Book definition (above) and our passing parameters (this line).  One important exception: parameters that have defaults <i>must</i> be the last.  Here we choose not to default to Jane Smith but to use Bill Nye.  If we skipped Bill Nye, the Jane Smith would have been used automatically.</p>
